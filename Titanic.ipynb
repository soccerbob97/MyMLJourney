```python
#filling in NAN values 
from sklearn.base import TransformerMixin
import sys
import numpy as np
np.set_printoptions(threshold=np.inf)

class AgeImputer(TransformerMixin):
    def __init__(self):
        #print("In titatnic imputer")
        self.menAverageAge = 0
        self.womenAverageAge = 0
        #self.embarked = embarked
    def fit(self,missing_titanic_df,y=None):
        self.maleAverageAge = (titanic_train[titanic_train["Sex"] == "male"])["Age"].mean()
        self.femaleAverageAge = (titanic_train[titanic_train["Sex"] == "female"])["Age"].mean()
        return self
    def transform(self,missing_titanic_df,y=None):
        if self.maleAverageAge == 0 and self.femaleAverageAge == 0:
            raise Exception("first call fit with training data")
        missing_titanic_df_copy = missing_titanic_df.copy()
        missing_titanic_df_copy.loc[missing_titanic_df_copy["Sex"] == "male","Age"] \
            = missing_titanic_df_copy.loc[missing_titanic_df_copy["Sex"] == "male","Age"].fillna(self.maleAverageAge)
        missing_titanic_df_copy.loc[missing_titanic_df_copy["Sex"] == "female","Age"] \
            = missing_titanic_df_copy.loc[missing_titanic_df_copy["Sex"] == "female","Age"].fillna(self.femaleAverageAge)
        #missing_titanic_df_copy["Embarked"].fillna(self.embarked,inplace=True)
        print('Age Imputer ')
        print(missing_titanic_df_copy)
        return missing_titanic_df_copy

"""
titanic_train_copy = titanic_train.copy()
imp = AgeImputer()
imp.fit(titanic_train_copy)
titanic_train_editted = imp.transform(titanic_train_copy)
"""



```

```python
from sklearn.base import BaseEstimator
class EmbarkedImputer(TransformerMixin,BaseEstimator):
    def __init__(self,embarked):
        self.embarked = embarked
    def fit(self,y=None):
        return self
    def transform(self,missing_titanic_df,y=None):
        missing_titanic_df_copy = missing_titanic_df.copy()
        missing_titanic_df_copy["Embarked"].fillna(self.embarked,inplace=True)
        print('Embarked Imputer')
        print(missing_titanic_df_copy)
        return missing_titanic_df_copy
        
```

```python
# Create new attributes from existing attributes 
from sklearn.preprocessing import OneHotEncoder

class AttributesTransformer(BaseEstimator,TransformerMixin):
    #def __init__(self,initial_titanic_dataset):
        #self.initial_titanic_dataset = initial_titanic_dataset
    def fit(self,missing_titanic_df,y=None):
        return self
    def transform(self,missing_titanic_df,y=None):
        missing_titanic_df_copy = missing_titanic_df.copy()
        missing_titanic_df_copy["Cabin"].fillna("",inplace=True)
        number_of_cabins_start = missing_titanic_df_copy["Cabin"].str.split(" ").str.len()
        #print(number_of_cabins_start)
        missing_titanic_df_copy["NumberOfCabins"] = np.where(missing_titanic_df["Cabin"].isna(),number_of_cabins_start-1,number_of_cabins_start)
        #print(missing_titanic_df_copy["NumberOfCabins"].to_string())
        missing_titanic_df_copy.drop(columns="Cabin",inplace=True)
        missing_titanic_df_copy["GenderNumber"] = missing_titanic_df_copy["Sex"].map({'female':0, 'male':1})
        missing_titanic_df_copy.drop(columns="Sex",inplace=True)
        ticketNumberColumn = missing_titanic_df_copy["Ticket"].str.split(" ")
        missing_titanic_df_copy["TicketNumber"] = np.where(ticketNumberColumn.str.len() > 1,ticketNumberColumn.str[-1],ticketNumberColumn.str[0])
        missing_titanic_df_copy["TicketNumber"].replace("LINE",0,inplace=True)
        missing_titanic_df_copy.drop(columns="Ticket",inplace=True)
        cat_encoder = OneHotEncoder()
        embarked = missing_titanic_df_copy[["Embarked"]]
        embarkedTransformed = cat_encoder.fit_transform(embarked)
        missing_titanic_df_copy["Cherbourg"] = embarkedTransformed[:,0].toarray()
        missing_titanic_df_copy["Queenstown"] = embarkedTransformed[:,1].toarray()
        missing_titanic_df_copy["Southampton"] = embarkedTransformed[:,2].toarray()
        missing_titanic_df_copy.drop(columns="Name",inplace=True)
        missing_titanic_df_copy.drop(columns="Embarked",inplace=True)
        print('Attributes Transformer')
        print(missing_titanic_df_copy)
        return missing_titanic_df_copy
"""
attributesTransformer = AttributesTransformer()
embarkedFiller = EmbarkedImputer(embarked="Q")
titanic_train_embarked_filled = embarkedFiller.transform(titanic_train_editted)
return_df = attributesTransformer.transform(titanic_train_embarked_filled)
print(return_df)
"""

    
```
