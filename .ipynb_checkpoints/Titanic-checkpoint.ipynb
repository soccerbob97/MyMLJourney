```python
#filling in NAN values for Age
from sklearn.base import TransformerMixin
import sys
import numpy as np
np.set_printoptions(threshold=np.inf)

class AgeImputer(TransformerMixin):
    def __init__(self):
        self.menAverageAge = 0
        self.womenAverageAge = 0
    def fit(self,missing_titanic_df,y=None):
        #Find average age for men and women
        self.maleAverageAge = (titanic_train[titanic_train["Sex"] == "male"])["Age"].mean()
        self.femaleAverageAge = (titanic_train[titanic_train["Sex"] == "female"])["Age"].mean()
        return self
    def transform(self,missing_titanic_df,y=None):
        if self.maleAverageAge == 0 and self.femaleAverageAge == 0:
            raise Exception("first call fit with training data")
        #fill in men and women age with appropriate data
        missing_titanic_df_copy = missing_titanic_df.copy()
        missing_titanic_df_copy.loc[missing_titanic_df_copy["Sex"] == "male","Age"] \
            = missing_titanic_df_copy.loc[missing_titanic_df_copy["Sex"] == "male","Age"].fillna(self.maleAverageAge)
        missing_titanic_df_copy.loc[missing_titanic_df_copy["Sex"] == "female","Age"] \
            = missing_titanic_df_copy.loc[missing_titanic_df_copy["Sex"] == "female","Age"].fillna(self.femaleAverageAge)
        return missing_titanic_df_copy


```

```python
# Custom Imputers that seperately fill in Nan embarked and fare values
from sklearn.base import BaseEstimator
class EmbarkedImputer(TransformerMixin,BaseEstimator):
    def __init__(self,embarked):
        self.embarked = embarked
    def fit(self,y=None):
        return self
    def transform(self,missing_titanic_df,y=None):
        missing_titanic_df_copy = missing_titanic_df.copy()
        missing_titanic_df_copy["Embarked"].fillna(self.embarked,inplace=True)
        return missing_titanic_df_copy
class FareImputer(TransformerMixin,BaseEstimator):
    def fit(self,missing_titanic_df,y=None):
        self.averageFare = missing_titanic_df["Fare"].mean()
        return self
    def transform(self,missing_titanic_df,y=None):
        missing_titanic_df_copy = missing_titanic_df.copy()
        missing_titanic_df_copy["Fare"].fillna(self.averageFare,inplace=True)
        return missing_titanic_df_copy
```

```python
# Create new attributes from existing attributes 
from sklearn.preprocessing import OneHotEncoder

class AttributesTransformer(BaseEstimator,TransformerMixin):
    #def __init__(self,initial_titanic_dataset):
        #self.initial_titanic_dataset = initial_titanic_dataset
    def fit(self,missing_titanic_df,y=None):
        return self
    def transform(self,missing_titanic_df,y=None):
        missing_titanic_df_copy = missing_titanic_df.copy()
        missing_titanic_df_copy["Cabin"].fillna("",inplace=True)
        number_of_cabins_start = missing_titanic_df_copy["Cabin"].str.split(" ").str.len()
        # If initial spot is Nan, we subtract 1 from cabin # because having a blank string results in len(array) = 1
        missing_titanic_df_copy["NumberOfCabins"] = np.where(missing_titanic_df["Cabin"].isna(),number_of_cabins_start-1,number_of_cabins_start)
        missing_titanic_df_copy.drop(columns="Cabin",inplace=True)
        missing_titanic_df_copy["GenderNumber"] = missing_titanic_df_copy["Sex"].map({'female':0, 'male':1})
        missing_titanic_df_copy.drop(columns="Sex",inplace=True)
        ticketNumberColumn = missing_titanic_df_copy["Ticket"].str.split(" ")
        missing_titanic_df_copy["TicketNumber"] = np.where(ticketNumberColumn.str.len() > 1,ticketNumberColumn.str[-1],ticketNumberColumn.str[0])
        missing_titanic_df_copy["TicketNumber"].replace("LINE",0,inplace=True)
        missing_titanic_df_copy.drop(columns="Ticket",inplace=True)
        cat_encoder = OneHotEncoder()
        embarked = missing_titanic_df_copy[["Embarked"]]
        embarkedTransformed = cat_encoder.fit_transform(embarked)
        missing_titanic_df_copy["Cherbourg"] = embarkedTransformed[:,0].toarray()
        missing_titanic_df_copy["Queenstown"] = embarkedTransformed[:,1].toarray()
        missing_titanic_df_copy["Southampton"] = embarkedTransformed[:,2].toarray()
        missing_titanic_df_copy.drop(columns="Name",inplace=True)
        missing_titanic_df_copy.drop(columns="Embarked",inplace=True)
        return missing_titanic_df_copy


    
```

```python
#transform initial data sets to competely filled data table
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

data_cleaning_pipeline = Pipeline([
     ('Age_Imputer', AgeImputer()),
     ('Fare_Imputer',FareImputer()),
     ("imputerEmbarked", EmbarkedImputer(embarked="S")),
     ('attributesTransformer', AttributesTransformer())
     ('std_scaler', StandardScaler()),
])

titanic_train_num = titanic_train.copy()
titanic_train_prepared = data_cleaning_pipeline.fit_transform(titanic_train_num)
titanic_test_prepared = data_cleaning_pipeline.transform(titanic_test)
titanic_final_test_prepared = data_cleaning_pipeline.transform(titanic_final_test)
```

```python
# Create helper functions that fit model, test model with test data, and output performance metrics
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict 
from sklearn.metrics import confusion_matrix

def confusion_matrix_helper(model,data,labels):
    survived_predictions = cross_val_predict(model,data,labels,cv=3)
    survived_predictions_processed = np.where(survived_predictions < .5,0,1)
    print('Confusion Matrix:')
    print(confusion_matrix(survived_predictions_processed,labels))


def train_model(model,training_data,training_labels):
    model.fit(training_data,training_labels)
    survived_predictions = model.predict(training_data)
    survived_predictions_processed = np.where(survived_predictions < .5,0,1)
    coefficient_determination_training = model.score(training_data,training_labels)
    print('coefficient of determination ',coefficient_determination_training)
    mse = mean_squared_error(training_labels,survived_predictions_processed)
    print('mse ',mse)
    rmse = np.sqrt(mse)
    print('rmse ',rmse)
    confusion_matrix_helper(model,training_data,training_labels)

def model_cross_validation(model,training_data,training_label):
    scores = cross_val_score(model, training_data, training_label,
                             scoring="neg_mean_squared_error", cv=10)
    rmse_scores = np.sqrt(-scores)
    print(rmse_scores)
    average_rmse_score = 0
    for num in rmse_scores:
        average_rmse_score += num
    average_rmse_score /= 10
    print(average_rmse_score)
    
def test_model(model,test_data,test_labels):
    test_predictions = model.predict(test_data)
    test_predictions_processed = np.where(test_predictions < .5, 0, 1)
    #seriesPd = pd.Series(lin_test_predictions_prepared)
    #print((titanic_test_labels.reset_index(drop=True) == seriesPd.reset_index(drop=True)).to_string())
    coefficent_determination_test = model.score(test_data,test_labels)
    print(coefficent_determination_test)
    test_mse = mean_squared_error(test_labels,test_predictions_processed)
    print('test mse ',test_mse)
    test_rmse = np.sqrt(test_mse)
    print('test rmse ',test_rmse)
    confusion_matrix_helper(model,test_data,test_labels)

def test_final(model,test_final_data):
    test_final_predictions = model.predict(test_final_data)
    test_predictions_processed = np.where(test_final_predictions < .5,0,1)
    final_df = pd.DataFrame()
    final_df["PassengerId"] = test_final_data["PassengerId"]
    final_df["Survived"] = pd.Series(test_predictions_processed)
    final_df.to_csv("final_test_survived.csv",index=False)
    
```

```python
# Train and Test Linear Regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

print('Linear Regression Training')
lin_reg = LinearRegression()
train_model(lin_reg,titanic_train_prepared,titanic_train_labels)
print('')

print('Linear Regression Cross validation')
model_cross_validation(lin_reg,titanic_train_prepared,titanic_train_labels)
print(' ')

print('Linear Regression testing')
test_model(lin_reg,titanic_test_prepared,titanic_test_labels)
print(' ')


```
